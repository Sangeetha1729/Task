```python
import world_bank_data as wb
import pandas as pd
import matplotlib.pyplot as plt
indicator = 'NY.GDP.PCAP.CD' 
countries = ['CD', 'BI']      
data = wb.get_series(indicator, country=countries, date='2000:2022', id_or_value='id')
df = data.unstack(level=0)
df.index = pd.to_datetime(df.index.get_level_values(1)).year
df.columns = ['DR Congo', 'Burundi']
df.plot(marker='o', title="GDP per capita (2000–2022)",
        ylabel="GDP per capita (current US$)", xlabel="Year", figsize=(10,5))
plt.xticks(ticks=range(2000, 2023, 2))
plt.grid(True)
plt.tight_layout()
plt.figtext(0.5, -0.05, "Source: World Bank via world_bank_data package", ha="center")
plt.savefig("gdp_per_capita_comparison.png", dpi=300)  
plt.show()
```
